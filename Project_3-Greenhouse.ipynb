```python
from tkinter import *
import joblib
window=Tk()
window.title('GreenHouse')
window.geometry('400x600')
l0=Label(window,text='Temperature')
l0.grid(row=0,column=1)
e0=Entry(window)
e0.grid(row=0,column=2)


l1=Label(window,text='Humidity')
l1.grid(row=1,column=1)
e1=Entry(window)
e1.grid(row=1,column=2)


l2=Label(window,text='MQ')
l2.grid(row=2,column=1)
e2=Entry(window)
e2.grid(row=2,column=2)


l3=Label(window,text='S1')
l3.grid(row=3,column=1)
e3=Entry(window)
e3.grid(row=3,column=2)


l4=Label(window,text='S2')
l4.grid(row=4,column=1)
e4=Entry(window)
e4.grid(row=4,column=2)


l5=Label(window,text='Result')
l5.grid(row=5,column=1)
e5=Entry(window)
e5.grid(row=5,column=2)


def clicked():
    print('button clicked')
    Temperature=e0.get()
    Humidity=e1.get()
    MQ=e2.get()
    S1=e3.get()
    S2=e4.get()
    print(Temperature,Humidity,MQ,S1,S2)
    model3=joblib.load('gh.joblib')
    p=model3.predict([[Temperature,Humidity,MQ,S1,S2]])
    print(p,type(p))
    if p[0]==0:
        e5.insert(0,'not ok')
    else:
        e5.insert(0,'ok')
    

b=Button(window,text='predict',command=clicked)
b.grid(row=12,column=1)
window.mainloop()
```
